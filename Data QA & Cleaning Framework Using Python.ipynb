```python
def standardize_emails(df):
    # Basic email normalization: lowercase and strip spaces
    df['customer_email'] = df['customer_email'].astype(str).str.lower().str.strip()
    # Mark obviously invalid emails
    df['email_valid'] = df['customer_email'].str.contains(r'^[\w\.-]+@[\w\.-]+\.[a-zA-Z]{2,}$', na=False)
    return df

def fill_missing_amounts(df):
    # Convert amount to numeric and fill missing with 0 (flagged later)
    df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
    df['amount_filled'] = df['amount'].fillna(0)
    df['amount_missing_flag'] = df['amount'].isna()
    return df

def standardize_status(df):
    df['status'] = df['status'].astype(str).str.lower().str.strip()
    return df

def run_all_cleaning(path):
    df = load_data(path)
    df = standardize_emails(df)
    df = fill_missing_amounts(df)
    df = standardize_status(df)
    return df

if __name__ == '__main__':
    import sys
    df = run_all_cleaning(sys.argv[1] if len(sys.argv)>1 else 'sample_data.csv')
    print(df.head())
```
